```python
response.status_code
```

```python
data = response.json()
data
```

### Convierto la api en un DataFrame

```python
def ranking_atp():
    results = data.get('results')
    rankings = results.get('rankings')
    df = pd.DataFrame(rankings)
    df_sort = df.sort_values(by='race_points', ascending=False, na_position='first')
    return df_sort
```

```python
ranking_atp()
```

```python
ranking_atp().to_csv('ranking_atp().csv')
```
