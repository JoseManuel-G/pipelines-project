## Análisis del DataFrame 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tenis = pd.read_csv('tennis.csv')
```

```python
tenis
```

```python
# Quiero saber quien es el jugador que más Grand Slams ha ganado de 2000 a 2016
# Para ello filtro el DataFrame para solo finales de Grand Slams y hago un value counts de la columna Winner
```

```python
def GS_0_16():
    tenis1 = tenis[tenis.Round=='The Final']
    tenis2 = tenis1[tenis1.Series=='Grand Slam']
    return tenis2
```

```python
GS_0_16()
```

```python
GS_0_16().Winner.value_counts()
```

```python
sns.countplot(y=GS_0_16().Winner)
```

#### Observamos que el resultado nos dice que el que más Grand Slams consigue en estas fechas es Federer.
#### SIn embargo como Federer empezó en el circuito ATP antes, filtramos este DataFrame por fechas, para ver quien ganó más Grand Slams desde que Djokovic Nadal están en el circuito ATP (año 2006) 

```python
data2 = GS_0_16().loc[17708:]  # Filtro por es fila porque es la fecha en la que empieza el primer torne de 2006
```

```python
data2
```

```python
data2.Winner.value_counts()
```

```python
sns.countplot(y=data2.Winner)
```

#### Aquí encambio observamos que el jugador que más Grand Slams ha ganado desde que los tres están en el circuito es Rafael Nadal.

```python
def Roger_gs_weak_era():   
    data3 = GS_0_16().loc[:17707] # Realizo esta función para saber cuantos Grand Slams ganó Federer antes de que sus rivales verdaderos comenzaran en el circuito ATP
    return data3.Winner.value_counts()
```

```python
Roger_gs_weak_era() 
```

#### Roger parte con una ventaja de 6 Grand Slams que ganó en la famosa weak era, ya que no tenía grandes rivales.

# Head to head Rafa vs Roger

```python
def wins_Rafa():
    tenis1 = tenis[tenis.Winner=='Nadal R.']
    tenis2 = tenis1[tenis1.Loser=='Federer R.']
    return tenis2.Winner.value_counts()
```

```python
wins_Rafa()
```

```python
def wins_Roger():
    tenis1 = tenis[tenis.Winner=='Federer R.']
    tenis2 = tenis1[tenis1.Loser=='Nadal R.']
    return tenis2.Winner.value_counts()
```

```python
wins_Roger()
```

#### Como podemos observar del total de partidos que han jugado, Rafa le ha ganado 23 y Roger solo 11.
